# 3. Drawing into Axes
All plots are drawn into axes. It is easiest to understand how matplotlib works if you use the object-oriented style.

```python
# Create some data to plot
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6 * x)
fig, ax = plt.subplots()
ax.plot(x, y)
fig, axes = plt.subplots(ncols=2, figsize=(8, 4))
ax0, ax1 = axes
ax0.plot(x, y)
ax1.plot(x, z)
```

# 4. Labelling Plots

```python
fig, axes = plt.subplots(figsize=(8, 4), ncols=2)
ax0, ax1 = axes

# First plot
ax0.plot(x, y)
ax0.set_xlabel("x")
ax0.set_ylabel("y")
ax0.set_title("x vs. y")

# Second plot
ax1.plot(x, z)
ax1.set_xlabel("x")
ax1.set_ylabel("z")
ax1.set_title("x vs. z")

# Squeeze everything in
fig.tight_layout()
```

## Line Styles

```python
fig, ax = plt.subplots(figsize=(12, 6), ncols=3)
ax[0].plot(x, y, linestyle="dashed")
ax[0].plot(x, z, linestyle="--")

ax[1].plot(x, y, linestyle="dotted")
ax[1].plot(x, z, linestyle=":")

ax[2].plot(x, y, linestyle="dashdot", linewidth=5)
ax[2].plot(x, z, linestyle="-.", linewidth=0.5)
```

## Colors

As described in the colors documentation, there are some special codes for commonly used colors:

-  b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

```python
fig, ax = plt.subplots()
ax.plot(x, y, color="k")
ax.plot(x, z, color="r")
```

```python
# Other ways to specify colors
fig, ax = plt.subplots(figsize=(16, 6), ncols=3)

# Grayscale
ax[0].plot(x, y, color="0.75")
ax[0].plot(x, z, color="0.2")

# RGB Tuple
ax[1].plot(x, y, color=(1, 0, 0.7))
ax[1].plot(x, z, color=(0, 0.4, 0.3))

# HTLM HEX Code
ax[2].plot(x, y, color="#00dcba")
ax[2].plot(x, z, color="#b029ba")
```

There is a default color cycle built into matplotlib.

```python
plt.rcParams["axes.prop_cycle"]
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
for factor in np.linspace(0.2, 1, 11):
    ax.plot(y, z * factor)
```

## Markers
There are lots of different markers availabile in matplotlib!


```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 5))

ax[0].plot(x[:20], y[:20], marker=".")
ax[0].plot(x[:20], z[:20], marker="o")

ax[1].plot(
    x[:30], z[:30], marker="^", markerfacecolor="r", markersize=10, markeredgecolor="k"
)
```

## Labels, Ticks and Gridlines


```python
fig, ax = plt.subplots(figsize=(24, 7), ncols=2)

ax[0].plot(x, y)

ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].set_title(r"A Complicated math function $f(x) = \cos(x)$")

ax[0].set_xticks(np.pi * np.array([-1, 0, 1]))
ax[0].set_xticklabels([r"$-\pi$", "0", r"$\pi$"])
ax[0].set_yticks([-1, 0, 1])

ax[0].set_yticks(np.arange(-1, 1.1, 0.2), minor=True)
ax[0].set_xticks(np.arange(-3, 3.1, 0.2), minor=False)

ax[0].grid(which="minor", linestyle="--")
ax[0].grid(which="major", linewidth=2)

ax[1].plot(x, y)

ax[1].set_xlabel("x")
ax[1].set_ylabel("y")
ax[1].set_title(r"A Complicated math function $f(x) = \cos(x)$")

ax[1].set_xticks(np.pi * np.array([-1, 0, 1]))
ax[1].set_xticklabels([r"$-\pi$", "0", r"$\pi$"])
ax[1].set_yticks([-1, 0, 1])

ax[1].set_yticks(np.arange(-1, 1.1, 0.2), minor=True)
ax[1].set_xticks(np.arange(-3, 3.1, 0.2), minor=True)

ax[1].grid(which="minor", linestyle="--")
ax[1].grid(which="major", linewidth=2)
```

```python
# Import necessary packages
import matplotlib.pyplot as plt
import pandas as pd

# Create data
boulder_precip = pd.DataFrame(
    columns=["date", "precip"],
    data=[
        ["2013-09-09", 0.1],
        ["2013-09-10", 1.0],
        ["2013-09-11", 2.3],
        ["2013-09-12", 9.8],
        ["2013-09-13", 1.9],
        ["2013-09-14", 0.01],
        ["2013-09-15", 1.4],
        ["2013-09-16", 0.4],
    ],
)
# Create plot
fig, ax = plt.subplots()
ax.bar(boulder_precip["date"].values, boulder_precip["precip"].values)
ax.set(
    title="Daily Precipitation (inches)\nBoulder, Colorado 2013",
    xlabel="Date",
    ylabel="Precipitation (Inches)",
)
plt.setp(ax.get_xticklabels(), rotation=45)
```
